```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf

# 1. Generate sample data (141 data points to match the 0-140 x-axis in your example)
t = np.arange(141)

# Create a smooth negative linear trend: y = -m*x + b
# We are using zero noise here to perfectly replicate the smooth ACF curve from class
ts_negative_trend = -50 * t + 7000 

# 2. Set up the figure layout (2 rows, 1 column)
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))

# Plot the Time Series
ax1.plot(t, ts_negative_trend)
ax1.set_title('Time Series with Negative Linear Trend')
ax1.set_xlabel('Time (t)')
ax1.set_ylabel('Value')

# Plot the ACF
# Setting lags=140 forces the plot to show the correlation across the entire dataset length
plot_acf(ts_negative_trend, ax=ax2, lags=140)
ax2.set_title('Autocorrelation')
ax2.set_xlabel('Lag')
ax2.set_ylabel('ACF')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf

# 1. Generate sample data
t = np.arange(141)

# Create a linear trend
trend = 0.5 * t

# Create seasonality (a sine wave repeating every 12 periods)
# Increasing the multiplier (10) makes the seasonal swings more drastic
seasonality = 10 * np.sin(2 * np.pi * t / 12)

# Combine them (we keep noise zero for a perfectly clean theoretical plot)
ts_trend_seasonality = trend + seasonality

# 2. Set up the figure layout (2 rows, 1 column)
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))

# Plot the Time Series
ax1.plot(t, ts_trend_seasonality)
ax1.set_title('Time Series with Trend and Seasonality')
ax1.set_xlabel('Time (t)')
ax1.set_ylabel('Value')

# Plot the ACF
# Setting lags=140 to see the full decay and repeated seasonal cycles
plot_acf(ts_trend_seasonality, ax=ax2, lags=140)
ax2.set_title('Autocorrelation')
ax2.set_xlabel('Lag')
ax2.set_ylabel('ACF')

plt.tight_layout()
plt.savefig('trend_seasonality_acf.png')
```
